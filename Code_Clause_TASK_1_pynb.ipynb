# **Detecting color blue**


It can detect any color but now we are detecting color blue


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/colors.png')
print('Input Image')
cv2_imshow(img)
Blue_Channel = img[:,:,0]
Green_Channel = img[:,:,1]
Red_Channel = img[:,:,2]
print('Blue Channel')
cv2_imshow(Blue_Channel)

grid_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
print('Gray Channel')
cv2_imshow(grid_gray)

grid_blue = cv2.subtract(Blue_Channel,grid_gray)
print('Subtracted Image')
cv2_imshow(grid_blue)

ret,thresh = cv2.threshold(grid_blue,110,255,cv2.THRESH_BINARY)
print('Threshold Image')
cv2_imshow(thresh)
```
